## Text Preprocessing and stop-word removal

```python
import re
rows,columns=df.shape

stopwords = [
    'a', 'about', 'above', 'after', 'again', 'against', 'all', 'am', 'an', 'and',
    'any', 'are', "aren't", 'as', 'at', 'be', 'because', 'been', 'before', 'being',
    'below', 'between', 'both', 'but', 'by', 'can', "can't", 'cannot', 'could',
    "couldn't", 'did', "didn't", 'do', 'does', "doesn't", 'doing', "don't", 'down',
    'during', 'each', 'few', 'for', 'from', 'further', 'had', "hadn't", 'has',
    "hasn't", 'have', "haven't", 'having', 'he', "he'd", "he'll", "he's", 'her',
    'here', "here's", 'hers', 'herself', 'him', 'himself', 'his', 'how', "how's",
    'i', "i'd", "i'll", "i'm", "i've", 'if', 'in', 'into', 'is', "isn't", 'it',
    "it's", 'its', 'itself', "let's", 'me', 'more', 'most', "mustn't", 'my',
    'myself', 'no', 'nor', 'not', 'of', 'off', 'on', 'once', 'only', 'or',
    'other', 'ought', 'our', 'ours', 'ourselves', 'out', 'over', 'own', "same",
    "shan't", 'she', "she'd", "she'll", "she's", 'should', "shouldn't", 'so',
    'some', 'such', 'than', "that", "that's", 'the', 'their', 'theirs', 'them',
    'themselves', 'then', 'there', "there's", 'these', 'they', "they'd", "they'll",
    "they're", "they've", 'this', 'those', 'through', 'to', 'too', 'under', 'until',
    'up', 'very', 'was', "wasn't", 'we', "we'd", "we'll", "we're", "we've", 'were',
    "weren't", 'what', "what's", 'when', "when's", 'where', "where's", 'which',
    'while', 'who', "who's", 'whom', "why", "why's", 'with', "won't", 'would',
    "wouldn't", 'you', "you'd", "you'll", "you're", "you've", 'your', 'yours',
    'yourself', 'yourselves',''
]


#create a new column
df['reason_for_recall_list'] = 'None'
df['reason_for_recall_cleaned'] = 'None'

for row_num in range(0,rows-1):
    #Remove all the special characters and numbers from the text
    df['reason_for_recall_cleaned'][row_num] = re.sub(r'[^a-zA-Z\s]', '', df['reason_for_recall'][row_num])
    
    #cleaning the sentence to all lower case and then splitting into words 
    df['reason_for_recall_list'][row_num]=' '.join([item for item in df['reason_for_recall_cleaned'][row_num].lower().split(' ') if item not in stopwords])
    #removing stop words

  
    
df.head()
```

## Tokenization

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer


# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the 'recall reasons' column
tfidf_matrix = tfidf_vectorizer.fit_transform(df['reason_for_recall_list'])

# Convert the TF-IDF matrix to a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Print the TF-IDF DataFrame
print(tfidf_df)
```

## Clustering

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

# Assume 'df' is your DataFrame with the tokenized and TF-IDF-transformed recall reasons
X = df['reason_for_recall_list']

# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X)

# Initialize K-Means with the desired number of clusters
num_clusters = 15  # Adjust as needed
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit K-Means to the TF-IDF data
kmeans.fit(X_tfidf)

# Get cluster assignments for each data point
cluster_labels = kmeans.labels_

# Add the cluster labels back to your DataFrame
df['cluster'] = cluster_labels
```
